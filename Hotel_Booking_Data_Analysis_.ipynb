#Data Analysis and Visualizations

```python
cancelled_perc= df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts() , edgecolor='k' , width=0.7)
plt.show()
```

```python
# To see which hotel have more cancellation rate/percentage
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df , palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hptels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['Not canceled','Canceled'])
plt.show()
```

```python
#calculate percentage of cancelled and not cancelled bookings in resort hotel
resort_hotel = df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#calculate percentage of cancelled and not cancelled bookings in city hotel
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#how prices are affecting on city and resort hotel  adr=(average daily rate)

resort_hotel =resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel =city_hotel.groupby('reservation_status_date')['adr'].mean()
```

```python


plt.figure(figsize = (10,5))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=10)
plt.plot(resort_hotel.index, resort_hotel.values, label='Resort Hotel') # use .values to access the values in the Series
plt.plot(city_hotel.index, city_hotel.values, label='City Hotel') # use .values to access the values in the Series
plt.legend(fontsize=10)
plt.show()
# it shows that the price of city hotels are less than resort hotel for some days and the spikes are for weekends.
```

```python
#To check in which month the reservations are more and cancelations

df['month'] = df['reservation_status_date'].dt.month_name()
plt.figure(figsize = (11,5))
ax1= sns.countplot(x = 'arrival_date_month', hue = 'is_canceled', data = df , palette = 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different months', size=20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['Not canceled','Canceled'])
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize=20)
#sns.barplot('month','adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
sns.barplot('month','adr', data = df.groupby('month')[['adr']].sum().reset_index())
#sns.barplot('month','adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
#which country have more cancelation only for top 10 countries
cancelled_data =df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(5,5))
plt.pie(top_10_country, autopct= '%.2f',labels = top_10_country.index)
plt.title('Top 10 countries with most cancellations')
plt.xlabel('Country')
```

```python
#from which platform the most of the custmers are cancelling the bookings
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
# adr/price is high for cancelled/not cancelled bookings
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')['adr'].mean()
cancelled_df_adr = cancelled_df_adr.reset_index()
cancelled_df_adr = cancelled_df_adr.sort_values('reservation_status_date')

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean()
not_cancelled_df_adr =not_cancelled_df_adr.reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date')

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate in Cancelled and Not Cancelled Bookings')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.legend()
plt.show()
```
